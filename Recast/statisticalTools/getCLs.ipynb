```python
from scipy.stats import chi2
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
ns = 1.0
nb = 0.5
nobs = 2.0
deltab = 0.5
chi2obs = np.vectorize(lambda mu: (nobs-nb-mu*ns)**2/deltab**2)
p = 1-chi2.cdf(chi2obs(1.0),df=1)
print(p)
```

```python
muvals = np.linspace(0.,3.,1000)
pvals = 1.0 - chi2.cdf(chi2obs(muvals),df=1)
```

```python
plt.plot(muvals,pvals)
plt.hlines(y=0.05,xmin=muvals.min(),xmax=muvals.max())
plt.xlim(2.4,2.5)
plt.grid()
plt.show()
```
